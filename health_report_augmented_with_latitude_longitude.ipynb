```python
import pandas as pd

df = pd.read_csv('health_report_pd.csv')
```

```python
import geonamescache
gc = geonamescache.GeonamesCache()
```

```python
def get_latitude_city(city, country_code):
    if gc.get_cities_by_name(city):
        return [list(x.values())[0] for x in gc.get_cities_by_name(city) if list(x.values())[0]['countrycode']==country_code][0]['latitude']
    else:
        return None

def get_longitude_city(city, country_code):
    if gc.get_cities_by_name(city):    
        return [list(x.values())[0] for x in gc.get_cities_by_name(city) if list(x.values())[0]['countrycode']==country_code][0]['longitude']
    else:
        return None
```

```python
df['country_code'] = df['countries'].apply(lambda x: gc.get_countries_by_names()[x]['iso'])
df.head(10)
```

```python
 df['latitude'] = df.apply(lambda x: get_latitude_city(x['cities'], x['country_code']), axis=1)
```

```python
 df['longitude'] = df.apply(lambda x: get_longitude_city(x['cities'], x['country_code']), axis=1)
```
